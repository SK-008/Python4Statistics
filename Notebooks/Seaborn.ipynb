```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Example Data
tips = sns.load_dataset('tips')
tips
```

```python
#Scatterplot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
#Scatterplots w/ Trendlines
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.title('Linear Trend Line')
plt.show()

sns.lmplot(x='total_bill', y='tip', data=tips, order=2)
plt.title('Polynomial Trend Line - 2nd Degree')
plt.show()

sns.lmplot(x='total_bill', y='tip', data=tips, ci=90)
plt.title('Linear Model Plot 90% Confidence Interval')
plt.show()

sns.lmplot(x='total_bill', y='tip', data=tips, ci=None)
plt.title('Linear Trend Line Without Confidence Interval')
plt.show()
```

```python
#Lineplot
sns.lineplot(x='size', y='total_bill', data=tips)
plt.show()
```

```python
#Barplot
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
#Histogram
sns.histplot(tips['total_bill'])
plt.show()
```

```python
#Boxplot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
#Pairplot - All variables plotted against each other
sns.pairplot(tips)
plt.show()
```

```python
#Faceting
g = sns.FacetGrid(tips, col='sex', row='time')
g.map(sns.scatterplot, 'total_bill', 'tip')
plt.show()

sns.lmplot(x='total_bill', y='tip', col='time', row='sex', data=tips)
plt.show()
```

```python
#Title & Label
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
#Adjusting the size of your graph
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```
