```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = 100
variance = 100

standard_deviation = np.sqrt(variance)

n1 = np.random.normal(loc=mean, scale=standard_deviation, size=100)
plt.hist(n1)
plt.show()

variance2 = 500
standard_deviation2 = np.sqrt(variance2)

n2 = np.random.normal(loc=mean, scale=standard_deviation2, size=100)

variance3 = 1000
standard_deviation3 = np.sqrt(variance3)

n3 = np.random.normal(loc=mean, scale=standard_deviation3, size=100)

print(n1)

plt.boxplot([n1, n2, n3], labels=['225 Variance', '450 Variance','675 Variance'])
```

```python
def calculate_iqr(data):
  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)
  iqr = q3 - q1
  print(iqr)

ns = [n1, n2, n3]

for i in ns:
    calculate_iqr(i)
```
