```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('affairs.csv')
df.head()
```

```python
y = df['hollingshead_occupation_code'] 
X = df[['have_kids','years_of_education']]
X = sm.add_constant(X)  
```

```python
model = sm.MNLogit(y, X)
result = model.fit()
result.summary()
```

```python
predictions = result.predict(X)
predictions
```
