```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
df = pd.read_csv('../Data/ts_s.csv', parse_dates=['Date'], index_col='Date')
df.head()
```

```python
y=df['Value']
y.plot(title='Time Series Data')
plt.show()
```

```python
#ADF Test for Stationarity 
adf_result = adfuller(y)
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
#y = np.diff(y)
adf_result = adfuller(y)
print('ADF Statistic:', adf_result[0])
print('p-value:', adf_result[1])
```

```python
plot_acf(y, lags=50)
plt.title('ACF')
plt.show()
plot_pacf(y, lags=50)
plt.title('PACF')
plt.show()
```

```python
#AR model
model = AutoReg(y, lags=1)
result = model.fit()
print(result.summary())
yhat = result.fittedvalues
e = result.resid
```

```python
#AR model

lags = list(range(1,10))

for i in lags:
    model = AutoReg(y, lags=i)
    result = model.fit()
    print(result.aic)
```

```python
#Fitted vs Actual
plt.figure(figsize=(12, 6))
plt.plot(y, label='Actual Values')
plt.plot(yhat, label='Fitted Values', color='red')
plt.title('Actual vs Fitted Values')
plt.legend()
plt.show()
```

```python
#Residual Plots
plt.plot(e)
plt.title('Residuals')
plt.show()

plot_acf(e, lags=40)
plt.title('ACF of Residuals')
plt.show()

sm.qqplot(e, line='s')
plt.title('Q-Q Plot')
plt.show()

plt.hist(e)
plt.title('Histogram of Residuals')
plt.show()
```

```python
# Forecasting
forecast_steps = 100  # Number of steps to forecast
forecast = result.predict(start=len(y), end=len(y) + forecast_steps - 1)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(y, label='Original Series')
plt.plot(pd.date_range(start=y.index[-1], periods=forecast_steps + 1, closed='right'), forecast, color='red', label='Forecast')
plt.title('Forecast')
plt.legend()
plt.show()
```
