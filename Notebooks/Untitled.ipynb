```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('affairs.csv')
df.head()
```

```python
y = df['number_of_affairs_last_year']
X = df[['is_male','age','years_married','has_children','religiousity_1to5','years_of_education','self_rating_of_marriage_1to5']]
X = sm.add_constant(X)
```

```python
#sns.pairplot(df)
plt.scatter(df['age'], df['number_of_affairs_last_year'])
```

```python
model = sm.OLS(y,X)
results = model.fit()
results.summary()
```
