# Solow-Swan Model - In Progress

```python
import matplotlib.pyplot as plt
import numpy as np

#Define parameters
alpha = 0.35  #Capital share of income
n = 0.02  #Population growth rate
g = 0.01  #Technological progress rate
delta = 0.05  #Depreciation rate

#Define time horizon and discretization
T = 50  # Number of periods
t = np.linspace(0, T, T + 1)  # Time points

#Define initial capital stock per worker
k0 = 2


def steady_state_capital():
  """Calculates steady state capital stock per worker."""
  return ((g + n) / (delta * alpha))**(1 / (1 - alpha))


def capital_dynamics(k, consumption):
  """Calculates capital stock dynamics for one period."""
  investment = np.maximum(0, np.power(k, alpha) - consumption)
  return (1 - delta) * k + investment


#Simulate capital stock path
k = np.zeros_like(t)
k[0] = k0

#Iterate through time periods
for i in range(1, T + 1):
  c = np.power(k[i - 1], 1 - alpha) * (1 - n - delta)  # Consumption
  k[i] = capital_dynamics(k[i - 1], c)

#Calculate and plot steady state capital
k_ss = steady_state_capital()
plt.plot(t, k, label="Capital Stock")
plt.axhline(y=k_ss, color='r', linestyle='--', label="Steady State")
plt.xlabel("Time")
plt.ylabel("Capital per Worker")
plt.title("Solow Growth Model Simulation")
plt.legend()
plt.grid(True)
plt.show()
```
