# Supply and Demand

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
#Parameters
a = 100
b = 2
c = 20
d = 1

#Demand function
def demand(P):
    return a - b * P

#Supply function
def supply(P):
    return c + d * P

#Equilibrium function
def equilibrium(P):
    return demand(P) - supply(P)

#Use fsolve to find the equilibrium price
P_eq = fsolve(equilibrium, x0=1)[0] #Initial guess of p=$1
Q_eq = demand(P_eq)

print(f'Equilibrium Price: {P_eq}')
print(f'Equilibrium Quantity: {Q_eq}')
```

```python
#Generate price range for plotting
prices = np.linspace(0, 80, 100)

#Calculate quantities for demand and supply
quantities_demand = demand(prices)
quantities_supply = supply(prices)

plt.figure(figsize=(10, 6))
plt.plot(prices, quantities_demand, label='Demand')
plt.plot(prices, quantities_supply, label='Supply')
#plt.axvline(P_eq, color='r', linestyle='--', label=f'Equilibrium Price: {P_eq:.2f}')
#plt.axhline(Q_eq, color='g', linestyle='--', label=f'Equilibrium Quantity: {Q_eq:.2f}')
plt.xlabel(f'Price - Equilibrium = ${P_eq:.2f}')
plt.ylabel(f'Quantity - Equilibrium = {Q_eq:.2f}')
plt.title('Supply and Demand Curves')
plt.legend()
plt.grid(True)
plt.show()
```
