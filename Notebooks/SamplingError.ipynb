```python
#Sampling Error
import numpy as np
import matplotlib.pyplot as plt

values = np.random.normal(loc=100, scale=15, size=1000000)

pop_mean = np.mean(values)
means = []

for i in range(5000):
    sample = np.random.choice(values, size=20)
    mean = sample.mean()
    means.append(mean)

plt.hist(means)
plt.title('Histogram of Sample Means (n=20, samples=5000)')
plt.axvline(pop_mean, color='red', linestyle='dashed', linewidth=2)
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()

#Highest and lowest mean 
highest_mean = max(means)
lowest_mean = min(means)
p25 = np.percentile(means, 25)  #25th percentile
p75 = np.percentile(means, 75)  #75th percentile



print(f'The highest sample mean was {lowest_mean} and the lowest was {highest_mean}')
```

```python
plt.hist(means)
plt.title('Histogram of Sample Means (n=20, samples=5000)')
plt.axvline(pop_mean, color='red', linewidth=2, label="Mean")
plt.axvline(highest_mean, color='purple', linewidth=2)
plt.axvline(lowest_mean, color='purple', linewidth=2)
plt.axvline(p25, color='green', linewidth=2)
plt.axvline(p75, color='green', linewidth=2)
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()
```
