```python
#Sampling Error
import numpy as np
import matplotlib.pyplot as plt

values = np.random.normal(loc=100, scale=15, size=1000000)

means = []

for i in range(5000):
    sample = np.random.choice(values, size=20)
    mean = sample.mean()
    means.append(mean)

plt.hist(means)
plt.title('Histogram of Sample Means (n=20, samples=5000)')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()

#Highest and lowest mean 
highest_mean = max(means)
lowest_mean = min(means)

print(f'The highest sample mean was {lowest_mean} and the lowest was {highest_mean}')
```
