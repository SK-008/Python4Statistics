# Bell Curves / Normally Distributed Data

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50, 50, 1000)
y1 = (1 / (5 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 0) / 5) ** 2)
y2 = (1 / (10 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 0) / 10) ** 2)
y3 = (1 / (15 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - 0) / 15) ** 2)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='σ=5')
plt.plot(x, y2, label='σ=10')
plt.plot(x, y3, label='σ=15')
plt.title('Bell Curves with Different Standard Deviations')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
