# Text Message Analysis - Frequency, Sentiment, & Response Time 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
df = pd.read_csv('/Users/robertwrobel/Downloads/tif2.csv')
```

```python
df['Message Date'] = pd.to_datetime(df['Message Date'])
df.set_index('Message Date', inplace=True)
df.head()
```

```python
df_inbound = df[df['Type'] == "Incoming"]
```

```python
#Frequency, average, and standard deviation of messages per period
messages_count_daily = df_inbound.groupby(pd.Grouper(freq='D')).size()
messages_count_weekly = df_inbound.groupby(pd.Grouper(freq='W')).size()
messages_count_monthly = df_inbound.groupby(pd.Grouper(freq='M')).size()

avg_messages_daily = df_inbound.groupby(pd.Grouper(freq='D')).size().mean()
avg_messages_weekly = df_inbound.groupby(pd.Grouper(freq='W')).size().mean()
avg_messages_monthly = df_inbound.groupby(pd.Grouper(freq='M')).size().mean()

std_messages_daily = df_inbound.groupby(pd.Grouper(freq='D')).size().std()
std_messages_weekly = df_inbound.groupby(pd.Grouper(freq='W')).size().std()
std_messages_monthly = df_inbound.groupby(pd.Grouper(freq='M')).size().std()

print(f"Average Number of Messages per Day: {avg_messages_daily:.2f}")
print(f"Average Number of Messages per Week: {avg_messages_weekly:.2f}")
print(f"Average Number of Messages per Month: {avg_messages_monthly:.2f}")
```

```python
#Message frequency plots
#plt.figure(figsize=(12, 6))
messages_count_daily.plot(kind='line', label='Daily')
plt.title('Daily Message Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

messages_count_weekly.plot(kind='line', label='Weekly')
plt.title('Weekly Message Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

messages_count_monthly.plot(kind='line', label='Monthly')
plt.title('Monthly Message Frequency')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Create sentiment analysis funciton
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

#sentiments = messages.apply(get_sentiment)
```

```python
#Group messages by period and concatenate texts
messages_daily = df_inbound.groupby(pd.Grouper(freq='D'))['Text'].apply(lambda x: ' '.join(x.dropna().astype(str)))
messages_weekly = df_inbound.groupby(pd.Grouper(freq='W'))['Text'].apply(lambda x: ' '.join(x.dropna().astype(str)))
messages_monthly = df_inbound.groupby(pd.Grouper(freq='M'))['Text'].apply(lambda x: ' '.join(x.dropna().astype(str)))
```

```python
#Apply sentiment analysis function
sentiments_daily = messages_daily.apply(get_sentiment)
sentiments_weekly = messages_weekly.apply(get_sentiment)
sentiments_monthly = messages_monthly.apply(get_sentiment)
```

```python
#Calculate average sentiment and standard deviation by period
avg_sentiment_daily = sentiments_daily.mean()
stdev_sentiment_daily = sentiments_daily.std()
avg_sentiment_weekly = sentiments_weekly.mean()
stdev_sentiment_weekly = sentiments_weekly.std()
avg_sentiment_monthly = sentiments_monthly.mean()
stdev_sentiment_monthly = sentiments_monthly.std()
```

```python
#Daily average sentiment plot 
sentiments_daily.plot(kind='line', label='Daily Average Composite Score')
plt.title('Text Message Sentiment Analysis')
plt.xlabel('Date')
plt.ylabel('Score')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Weekly average sentiment plot
sentiments_weekly.plot(kind='line', label='Weekly Average Composite Score')
plt.title('Text Message Sentiment Analysis')
plt.xlabel('Date')
plt.ylabel('Score')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Monthly average sentiment plot
sentiments_monthly.plot(kind='line', label='Monthly Average Composite Score')
plt.title('Text Message Sentiment Analysis')
plt.xlabel('Date')
plt.ylabel('Score')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Calculate response times
df = pd.read_csv('/Users/robertwrobel/Downloads/BM.csv')
```

```python
df['Message Date'] = pd.to_datetime(df['Message Date'])
#df.set_index('Message Date', inplace=True)
```

```python
df['Response Time'] = df['Message Date'].shift(-1) - df['Message Date']
df['Response Type'] = df['Type'].shift(-1)

#Only consider response times where the original message is outgoing and the response is incoming
response_times = df[(df['Type'] == 'Outgoing') & (df['Response Type'] == 'Incoming')]

daily_avg_response_time = response_times.resample('D', on='Message Date')['Response Time'].mean()
weekly_avg_response_time = response_times.resample('W', on='Message Date')['Response Time'].mean()
monthly_avg_response_time = response_times.resample('M', on='Message Date')['Response Time'].mean()

print("Daily Average Response Time:")
print(daily_avg_response_time)

print("\nWeekly Average Response Time:")
print(weekly_avg_response_time)

print("\nMonthly Average Response Time:")
print(monthly_avg_response_time)
```

```python
#Plot the response times
plt.figure(figsize=(12, 6))
daily_avg_response_time.plot(kind='line', label='Daily')
plt.title('Daily Average Response Time')
plt.xlabel('Date')
plt.ylabel('Average Response Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
weekly_avg_response_time.plot(kind='line', label='Weekly')
plt.title('Weekly Average Response Time')
plt.xlabel('Date')
plt.ylabel('Average Response Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
monthly_avg_response_time.plot(kind='line', label='Monthly')
plt.title('Monthly Average Response Time')
plt.xlabel('Date')
plt.ylabel('Average Response Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
