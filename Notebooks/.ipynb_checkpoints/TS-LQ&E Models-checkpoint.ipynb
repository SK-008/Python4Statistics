```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('ts_sandq.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head()
```

```python
plt.plot(df.index, df['Value'], label='Value')
plt.title('Time Series w/ Seasonality and Quadratic Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
#Linear Trend Model
df['Linear_Time_Trend']= np.arange(len(df))
print(df['Linear_Time_Trend'].head())
model = smf.ols('Value ~ 1 + Linear_Time_Trend', data=df).fit()
print(model.summary())
df['yhat'] = model.fittedvalues

plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['yhat'], label='Fitted Values', color='red', linestyle='--')
plt.title('Original Data and Fitted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
df['Quadratic_Time_Trend'] = df['Linear_Time_Trend'] * df['Linear_Time_Trend']
print(df['Quadratic_Time_Trend'].head())
model = smf.ols('Value ~ 1 + Linear_Time_Trend + Quadratic_Time_Trend', data=df).fit()
print(model.summary())
df['yhat'] = model.fittedvalues

plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['yhat'], label='Fitted Values', color='red', linestyle='--')
plt.title('Original Data and Fitted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Exponential Trend Model
df['Log_Value'] = np.log(df['Value'])
df['Log_Value'].head()
model = smf.ols('Log_Value ~ 1 + Linear_Time_Trend', data=df).fit()
print(model.summary())
df['yhat'] = model.fittedvalues

plt.subplot(2, 1, 1)
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['yhat'], label='Fitted Values', color='red', linestyle='--')
plt.title('Original Data and Fitted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Exponential Trend Model
df['Log_Value'] = np.log(df['Value'])
df['Log_Value'].head()
model = smf.ols('Log_Value ~ 1 + Linear_Time_Trend', data=df).fit()
print(model.summary())
df['yhat3'] = model.fittedvalues

plt.subplot(2, 1, 1)
plt.plot(df.index, df['Value'], label='Original Data')
plt.plot(df.index, df['yhat3'], label='Fitted Values', color='red', linestyle='--')
plt.title('Original Data and Fitted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
#Selecting Between Models Using Differences
df['Value_First_Diff'] = df['Value'].diff()
df['Value_Second_Diff'] = df['Value_First_Diff'].diff()
df['Value_Percent_Diff'] = df['Value'].pct_change()
```

```python
#plt.figure(figsize=(14, 10))

plt.subplot(4, 1, 1)
plt.plot(df.index, df['Value'], label='Original Data')
plt.title('Original Data')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(df.index, df['Value_First_Diff'], label='First Differences')
plt.title('First Differences')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(df.index, df['Value_Second_Diff'], label='Second Differences')
plt.title('Second Differences')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(df.index, df['Value_Percent_Diff'], label='Percent Differences')
plt.title('Percent Differences')
plt.legend()

plt.tight_layout()
plt.show()
```
