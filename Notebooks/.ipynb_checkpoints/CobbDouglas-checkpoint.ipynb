```python
import numpy as np
from scipy.optimize import minimize
```

```python
#Production parameters
#Parameters
A = 1
alpha = 0.3
beta = 0.7

#Cost parameters
C_max = 100  # Maximum cost
C_K = 5      # Cost of capital
C_L = 2      # Cost of labor
```

```python
#Cobb-Douglas production function: Q = A * K^alpha* L^beta
#A = TFP, K = capital, L = labor
def cobb_douglas(inputs, A, alpha, beta):
    K, L = inputs
    return -A * K**alpha * L**beta  #Minimize negative output to maximize output
```

```python
#Cost constraint function
def cost_constraint(inputs, C_max, C_K, C_L):
    K, L = inputs
    return C_K * K + C_L * L - C_max
```

```python
#Constraints
constraints = [{'type': 'ineq', 'fun': cost_constraint, 'args': (C_max, C_K, C_L)}]
```

```python
#Initial guess for inputs (capital and labor)
initial_guess = [10, 10]
```

```python
#Maximize the production function subject to the cost constraint
result = minimize(cobb_douglas, initial_guess, args=(A, alpha, beta), constraints=constraints)

#Extract the optimal inputs and output
optimal_inputs = result.x
optimal_output = -result.fun  # Convert back from minimized value to output value

print(f"Optimal inputs (K, L): {optimal_inputs}")
print(f"Optimal output: {optimal_output}")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Generate data for plotting
K = np.linspace(1, 20, 100)
L = np.linspace(1, 20, 100)
K, L = np.meshgrid(K, L)
Z = cobb_douglas([K, L], A, alpha, beta)

#Plot the 3D surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(K, L, Z, cmap='viridis')
ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labor (L)')
ax.set_zlabel('Output')
ax.set_title('Cobb-Douglas Production Function')
plt.show()
```
