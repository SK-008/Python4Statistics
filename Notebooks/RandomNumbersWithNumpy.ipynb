```python
#Generating Data with Numpy
import numpy as np
import pandas as pd
```

```python
#Set a random seed
np.random.seed(1663)
```

```python
#Uniform Distribution

#Generate 10 integers between 0 and 1 
x = np.random.rand(10) 
print(x)

#Generate 10 integers between 0 and 100 
x = np.random.randint(100, size=10)
print(x)

#Generate 10 integers between 50 and 100 
x = np.random.randint(50, 100, size=10)
print(x)

#Generate 2 sets of 10 values between 0 and 100
x = np.random.randint(100, size=(10,2))
print(x)

#Generate 10 floats between 0 and 100 
u = np.random.uniform(low=0, high=100, size=10)
print(u)
```

```python
#Generate 10 floats from the normal distribution
#loc = mean, scale = standard deviation, size = sample size 
n = np.random.normal(loc=0, scale=1, size=10)
print(n)
```

```python
#Generate 10 integers from the binomial distribution
#For binomial, n = number of trials, p = probability of success for a given trial, size = sample size
b = np.random.binomial(n=50, p=0.25, size=10)
print(b)
```

```python
#Generate 10 integers from the poisson distribution
#lam = average rate of occurrance, size = sample size
p = np.random.poisson(lam=5, size=10)
print(p)
```

```python
#Generate 10 observations from the gamma distribution
g = np.random.gamma(shape=2.0, scale=2.0, size=10)
g
```

```python
#Exponential Dist
e = np.random.exponential(scale=1.0, size=5)  #scale=mean
e
```

```python
#Create span of numbers ranging from the lowest value of a series to the highest, incremented evenly along n observations
z = [3, 6, 6, 8, 7, 7, 5, 0, 5, 6]
span = np.linspace(min(z), max(z), len(z))
print(span)
```

```python
#Select 3 items randomly from a list w/ and w/o replacement
choice = np.random.choice(z, size=3) 
print(choice)

choice = np.random.choice(z, size=3, replace=True) 
print(choice)
```

```python
#Shuffle / Permute
permuted = np.random.permutation(z)
permuted
```
